# **Feature Engineering**

---



```python
# --- Make a copy to avoid changing the original raw data ---
df_eng = weather_data.copy()

# --- 1. Date-Based Features ---
# First, ensure the 'Date' column is in datetime format
df_eng['Date'] = pd.to_datetime(df_eng['Date'])

# Extract year, month, and day
df_eng['Year'] = df_eng['Date'].dt.year
df_eng['Month'] = df_eng['Date'].dt.month
df_eng['Day'] = df_eng['Date'].dt.day

# It's often useful to know the day of the year as well
df_eng['DayOfYear'] = df_eng['Date'].dt.dayofyear

print("Created date-based features: Year, Month, Day, DayOfYear")
```

```python
# Daily temperature range
df_eng['TempRange'] = df_eng['MaxTemp'] - df_eng['MinTemp']

# Average temperature for the day
df_eng['AvgTemp'] = (df_eng['MinTemp'] + df_eng['MaxTemp']) / 2

print("Created temperature-based features: TempRange, AvgTemp")
```

```python
df_eng['PressureChange'] = df_eng['Pressure3pm'] - df_eng['Pressure9am']

print("Created pressure-based feature: PressureChange")
```

```python
df_eng['AvgWindSpeed'] = (df_eng['WindSpeed9am'] + df_eng['WindSpeed3pm']) / 2

print("Created wind-based feature: AvgWindSpeed")
```

```python
df_eng['HumidityTemp_Interaction'] = df_eng['AvgTemp'] * (df_eng['Humidity3pm'] * 0.01)

print("Created interaction feature: HumidityTemp_Interaction")
```

```python
df_eng = df_eng.drop('Date', axis=1)
print("\nDropped the original 'Date' column.")
```

```python
# --- Display the results ---
print("\n--- DataFrame with New Features (first 5 rows) ---")
display(df_eng.head())

print(f"\nOriginal number of columns: {len(weather_data.columns)}")
print(f"New number of columns: {len(df_eng.columns)}")
```

```python
print("\n--- Preparing 'RainToday' Feature ---")
if 'RainToday' in df_eng.columns and df_eng['RainToday'].dtype == 'object':
    df_eng['RainToday'] = df_eng['RainToday'].map({'No': 0, 'Yes': 1})
    print("Converted 'RainToday' to 0s and 1s. It will be treated as CATEGORICAL in the pipeline.")

print("\n--- Final DataFrame Info ---")
df_eng.info()
```

# **Pipeline**

---



```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier

# We assume 'X_train' is your training DataFrame, ready to be used.

# --- 1. Identify Numerical and Categorical Columns ---
numerical_features = X_train.select_dtypes(include=np.number).columns.tolist()
categorical_features = X_train.select_dtypes(exclude=np.number).columns.tolist()

# Move 'RainToday' to the categorical list
if 'RainToday' in numerical_features:
    numerical_features.remove('RainToday')
    categorical_features.append('RainToday')


print(f"Identified {len(numerical_features)} numerical features.")
print(f"Identified {len(categorical_features)} categorical features.")


# --- 2. Define the Preprocessing Steps for Each Column Type ---
# This is the same robust preprocessor we have used for all models.
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler()) # Kept for consistency and future stacking
])

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])


# --- 3. Create the Master Preprocessor ---
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ],
    remainder='drop'
)


# --- 4. Create the Final Decision Tree Pipeline ---
# This chains our preprocessor with the DecisionTreeClassifier.
dt_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),

    ('classifier', DecisionTreeClassifier(
        random_state=42
    ))
])

print("\nâœ… Full Decision Tree pipeline created successfully.")
print("\nPipeline Steps:")
display(dt_pipeline)
```

# **Evaluation and Selection**

# **Helper Function**

---



```python
def dt_hyperparameter(param_name, param_value):

    params = {param_name: param_value}

    dt_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', DecisionTreeClassifier(random_state=42, **params))
    ])

    # Train the pipeline
    dt_pipeline.fit(X_train, y_train)

    train_preds = dt_pipeline.predict(X_train)
    train_f1 = f1_score(y_train, train_preds)

    # --- Validation Metrics ---
    val_preds = dt_pipeline.predict(X_val)
    val_f1 = f1_score(y_val, val_preds)

    # Return a dictionary with the F1 Error for both sets.
    return {
         param_name: param_value,
        'train_f1_error': round(1 - train_f1, 4),
        'validation_f1_error': round(1 - val_f1, 4)
    }
```

# **RandomizedSearch CV**

---



```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# --- 1. Define the Hyperparameter Search Space ---
# We are defining a "distribution" for each hyperparameter.
# RandomizedSearchCV will randomly sample values from these distributions.
param_dist = {
    # 'classifier__' is the prefix for parameters of the classifier step in our pipeline.

    # Test both 'gini' and 'entropy'
    'classifier__criterion': ['gini', 'entropy'],

    # Search depths around our best manual finding (6), from simple to more complex.
    'classifier__max_depth': randint(low=3, high=15),

    # Search leaf samples around our best manual finding (200).
    # We use a wider range to let the search explore.
    'classifier__min_samples_leaf': randint(low=50, high=300),

    # Search split samples around our best manual finding (1000).
    'classifier__min_samples_split': randint(low=200, high=1500)
}


# --- 2. Set up RandomizedSearchCV ---
# n_iter=50 means it will run 50 different experiments (50 random combinations).
# cv=3 means each experiment will use 3-fold cross-validation.
# scoring='f1' tells it to optimize for the F1-score.
# n_jobs=-1 uses all available CPU cores to speed up the search.
random_search = RandomizedSearchCV(
    estimator=dt_pipeline,
    param_distributions=param_dist,
    n_iter=50,
    cv=3,
    scoring='f1',
    n_jobs=-1,
    random_state=42,
    verbose=1 # This will print progress updates as it runs.
)


# --- 3. Run the Search ---
print("--- Running RandomizedSearchCV for Decision Tree ---")
random_search.fit(X_train, y_train)


# --- 4. Report the Best Results ---
print("\nâœ… RandomizedSearchCV Complete.")
print(f"\nBest average F1-Score during cross-validation: {random_search.best_score_:.4f}")
print("\nBest Hyperparameters found:")
print(random_search.best_params_)

# You can also view the top 5 results in a DataFrame for more detail
cv_results_df = pd.DataFrame(random_search.cv_results_)
top_5_df = cv_results_df.sort_values(by='rank_test_score').head(5)
print("\n--- Detailed CV Results (Top 5) ---")
display(top_5_df[['param_classifier__criterion', 'param_classifier__max_depth',
                   'param_classifier__min_samples_leaf', 'param_classifier__min_samples_split',
                   'mean_test_score', 'std_test_score', 'rank_test_score']])
```

# **Evaluation**

---



```python
from sklearn.metrics import f1_score, roc_auc_score, classification_report, confusion_matrix
import plotly.figure_factory as ff

# --- Step 1: Extract the best model from the completed search ---
best_dt_model = random_search.best_estimator_
print("âœ… Best Decision Tree model extracted from RandomizedSearchCV.")
print(f"The best model has parameters: {random_search.best_params_}")


# --- Step 2: Make predictions on the validation set ---
print("\nMaking predictions on the unseen validation data...")
val_preds = best_dt_model.predict(X_val)
val_preds_proba = best_dt_model.predict_proba(X_val)[:, 1]


# --- Step 3: Calculate and print the final scores ---
final_f1_score = f1_score(y_val, val_preds)
final_auc_score = roc_auc_score(y_val, val_preds_proba)

print("\n--- Official Gauntlet Score for Tuned Decision Tree ---")
print(f"Validation F1-Score: {final_f1_score:.4f}")
print(f"Validation AUC-ROC:  {final_auc_score:.4f}")
print("-------------------------------------------------------")


# --- Step 4: Display the full Classification Report ---
print("\n--- Full Classification Report ---")
# The 'target_names' argument makes the report easier to read.
report = classification_report(y_val, val_preds, target_names=['No Rain', 'Rain'])
print(report)


# --- Step 5: Display the Confusion Matrix using Plotly ---
print("\n--- Confusion Matrix ---")
# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_val, val_preds)

# Create the Plotly figure
# We reverse the matrix and labels for the standard layout (TN, FP, FN, TP)
fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=['Predicted No Rain', 'Predicted Rain'],
    y=['Actual No Rain', 'Actual Rain'],
    colorscale='Blues'
)

fig.update_layout(title_text='Confusion Matrix')
fig.show()
```

# **Deployment**

```python
import gradio as gr

# --- Suppress Warnings ---
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. Load the Model ---
# This is the ONLY file we need to load.
MODEL_FILENAME = '04_decision_tree_tuned.joblib'
try:
    model_pipeline = joblib.load(MODEL_FILENAME)
    print(f"âœ… Model '{MODEL_FILENAME}' loaded successfully.")
except FileNotFoundError:
    print(f"âŒ FATAL ERROR: Model file '{MODEL_FILENAME}' not found.")
    print("--- Please ensure the model file is in the same directory as this script. ---")
    exit()

# --- 2. Define All Necessary Data Directly in the Script ---

# The exact list of unique locations, hard-coded for simplicity.
AUSTRALIA_LOCATIONS = [
    'Albury', 'BadgerysCreek', 'Cobar', 'CoffsHarbour', 'Moree', 'Newcastle',
    'NorahHead', 'NorfolkIsland', 'Penrith', 'Richmond', 'Sydney', 'SydneyAirport',
    'WaggaWagga', 'Williamtown', 'Wollongong', 'Canberra', 'Tuggeranong',
    'MountGinini', 'Ballarat', 'Bendigo', 'Sale', 'MelbourneAirport', 'Melbourne',
    'Mildura', 'Nhil', 'Portland', 'Watsonia', 'Dartmoor', 'Brisbane', 'Cairns',
    'GoldCoast', 'Townsville', 'Adelaide', 'MountGambier', 'Nuriootpa', 'Woomera',
    'Albany', 'Witchcliffe', 'PearceRAAF', 'PerthAirport', 'Perth', 'SalmonGums',
    'Walpole', 'Hobart', 'Launceston', 'AliceSprings', 'Darwin', 'Katherine', 'Uluru'
]

# The exact list of 30 columns the model was trained on.
MODEL_INPUT_COLUMNS = [
    'Location', 'MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
    'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm', 'WindSpeed9am',
    'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm',
    'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RainToday', 'Year', 'Month',
    'Day', 'DayOfYear', 'TempRange', 'AvgTemp', 'PressureChange', 'AvgWindSpeed',
    'HumidityTemp_Interaction'
]

# --- 3. Define the Prediction Function ---
def predict_rainfall(location, min_temp, max_temp, rainfall, wind_gust_speed, humidity_3pm, pressure_3pm, rain_today):
    input_data = {col: np.nan for col in MODEL_INPUT_COLUMNS}
    input_data.update({
        'Location': location, 'MinTemp': min_temp, 'MaxTemp': max_temp,
        'Rainfall': rainfall, 'WindGustSpeed': wind_gust_speed,
        'Humidity3pm': humidity_3pm, 'Pressure3pm': pressure_3pm,
        'RainToday': 1 if rain_today == 'Yes' else 0
    })
    input_df = pd.DataFrame([input_data])[MODEL_INPUT_COLUMNS]
    prediction_proba = model_pipeline.predict_proba(input_df)[0]
    confidences = {'No': prediction_proba[0], 'Yes': prediction_proba[1]}
    return confidences

# --- 4. Define the Gradio Interface ---
app_interface = gr.Interface(
    fn=predict_rainfall,
    title="ðŸŒ§ï¸ Australian Rain Predictor (Decision Tree Model)",
    description="Enter weather conditions to predict rain tomorrow. Powered by a tuned Decision Tree.",
    inputs=[
        gr.Dropdown(choices=AUSTRALIA_LOCATIONS, label="Location", value="Canberra"),
        gr.Number(label="Min Temperature (Â°C)", value=12.0, minimum=-8.5, maximum=33.9),
        gr.Number(label="Max Temperature (Â°C)", value=23.0, minimum=-4.8, maximum=48.1),
        gr.Number(label="Rainfall Today (mm)", value=0.0, minimum=0.0, maximum=371.0),
        gr.Number(label="Max Wind Speed (km/h)", value=40.0, minimum=6.0, maximum=135.0),
        gr.Number(label="Humidity at 3pm (%)", value=52.0, minimum=0.0, maximum=100.0),
        gr.Number(label="Pressure at 3pm (hPa)", value=1015.0, minimum=977.1, maximum=1039.6),
        gr.Radio(choices=['No', 'Yes'], label="Did it Rain Today?", value='No')
    ],
    outputs=gr.Label(num_top_classes=2, label="Probability of Rain Tomorrow"),
    allow_flagging='never'
)

# --- 5. Launch the Application ---
if __name__ == "__main__":
    app_interface.launch()
```
